```python
DAin = 0.27 #place
DAout = 0.3
BankMeasured = 20
bankBuffer=BankMeasured/2


DA2to10A = [5, 5, 0]
DA10to40 = [10, 10, 10]
DA40to160 = [30, 30, 10]
DA160to640 =[60, 60, 20]
DA640plus = [80, 80, 20]

acrespermile = 640
DA_AcresIn = DAin*acrespermile
DA_AcresOut = DAout*acrespermile
print(DA_AcresIn)
Areas = [DA_AcresIn]#, DA_AcresOut]
for area in Areas:  #an improvement is to test across the breaks to see if the setbacks change within the project area.  Look for DA near breakpoints or areas with a confluence.
    if area in range(0, 2):
        print("setbacks do not apply for drainage area less than 2 acres")
    if area in range(2, 10):
        setbacks = DA2to10A
    if area in range(2, 40):
        setbacks = DA10to40
    if area in range(40, 160):
        setbacks = DA40to160
    if area in range(160, 640):
        setbacks = DA160to640
    if area > 640:
        setbacks = DA640plus
    print(setbacks)
    print('Measured Bank Half Width ', bankBuffer, " feet")
    arcpy.analysis.Buffer(r"Stormwater Viewer\Stormwater Line\Open Channel", r"MeasuredBank", bankBuffer, "FULL", "ROUND", "ALL", None, "PLANAR")
    print('Default Bank Half Width ', setbacks[0]/2, " feet")
    arcpy.analysis.Buffer(r"Stormwater Viewer\Stormwater Line\Open Channel", r"DefaultBank", setbacks[0]/2, "FULL", "ROUND", "ALL", None, "PLANAR")
    print("Preservation setbacks at ", setbacks[1],  " feet")
    arcpy.analysis.Buffer(r"MeasuredBank", r"PreservationSetbackmeasured", setbacks[1], "FULL", "ROUND", "ALL", None, "PLANAR")
    arcpy.analysis.Buffer(r"DefaultBank", r"PreservationSetbackdefault", setbacks[1], "FULL", "ROUND", "ALL", None, "PLANAR")
    print("limited use setbacks at ", setbacks[2], " feet")
    arcpy.analysis.Buffer(r"PreservationSetbackmeasured", r"LimitedUseSetbackmeasured", setbacks[2], "FULL", "ROUND", "ALL", None, "PLANAR")
    arcpy.analysis.Buffer(r"PreservationSetbackdefault", r"LimitedUseSetbackdefault", setbacks[2], "FULL", "ROUND", "ALL", None, "PLANAR")




```
