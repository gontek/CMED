```python
#Ash Street Uplift Check 1
#Ash Street uplift @ STM-EX

#STM-EX approx 45 ft landward of the levee crest 
#Kansas River Unit Section II Sta 364+80 Sheet #3 North Lawrence Unit Section II 
#Kansas River FEMA cross section L 
#100 yr BFE = 823 ft  
#500 yr WSEL = 831.25 ft
#Levee Flood Protection Elevation from Lidar = 829.4 ft
#levee FPE from as built  = 829.0 ft

Gw = 62.4 #unit weight of water, 62.4 pcf

#assume pressure is dissipated to 0 at the ground surface
#Gw= #equivalent unit weight - (H1/H2)*Gw

RCP = 140 #pounds per ft weight of 15" RCP
#DesignElevations
FPE=829.4 #ks dasc tile 81683 2021 lidar, top of levee
F500YR = 831.25
F100YR = 823 # base flood elevation in feet
IBE = 810.0 #impervious blanket elevation in feet  #The top layers are sandy, silty clay at 364+80
GE = 824.0  #grade elevation in feet
FB = 812.2  #base of structure  elevation in feet structure STM-EX

w = 62.4 #pcf density of water
Conc = 150 #pcf density of concrete
RCP15 = 140  # plf of 15" RCP
RCP18 = 170 #plf of 18" RCP
Clay = 120 #lbs/ft3
Gravel = 100 #lbs/ft3
PPP24 = 11 #plf of 24" PP Pipe
PPP18 = 8 #plf pf 18" PP Pipe
    
#uplift Calculations for Trench Footings
def CASE_A(footing): #'river at top of flood protection structure'
    FS = 1.1
    HGA = FPE
    H1A = HGA-IBE
    H2A = GE-IBE
    H3A = GE-footing
    UPLIFTA = H3A*(H1A/H2A)*w*FS
    print(round(UPLIFTA, 1), "lb Uplift case A") 
    return UPLIFTA

def CASE_B(footing): #'design river stage @500 yr'
    FS = 1.5
    HGB = F500YR
    H1B = HGB-IBE
    H2B = GE-IBE
    H3B = GE-footing
    UPLIFTB = H3B*(H1B/H2B)*w*FS
    print(round(UPLIFTB, 1), "lb Uplift case B")
    return UPLIFTB
def CASE_C(footing): #'design river stage @500 yr'
    FS = 1.3
    HGC = F100YR
    H1C = HGC-IBE
    H2C = GE-IBE
    H3C = GE-footing
    UPLIFTC = H3C*(H1C/H2C)*w*FS
    print(round(UPLIFTC, 1), "lb Uplift C")
    return UPLIFTC

CASE_A(FB)
CASE_B(FB)
CASE_C(FB)

governingCase = (max(CASE_A(FB), CASE_B(FB), CASE_C(FB)))

print(round(governingCase, 2), "lbf max uplift pressure of governing case")

HGB = F500YR
H1B = HGB-IBE
H2B = GE-IBE
H3B = GE-FB
P1 = H1B*w
P2 = H2B*H1B/H2B*w
P3 = H3B*H1B/H2B*w

Uplift = P3

print ("P1:", round(P1, 2))
print ("P2:", round(P2, 2))
print ("P3:", round(P3, 2))
print("Uplift pressure:", round(Uplift, 2), " psf")

#Flotation Equation @ STM-EX
#(We+Wc+S)/(U-Wg) > SF
STM_EX = 3*3 #sq ft area
Ws = Conc*3*3*.5 + Conc*5*3*4 #weight of structure base + 4x 5 ft x 3 ft walls
Wc = 0#Gw*5*4 #weight of water in structure 5.2 ft deep * 2 ft x 2ft 
S = 0 #Surcharge Load 
Wg = w*(10.4-3)*2 #weight of surcharge water above top of structure
H = 4 #Head of water height over structure
U = P3*STM_EX #Uplift Force  = water density * Head of water height over structure

Result = (Ws+Wc+S)/(U-Wg)
SF= 1.5 #Normal Operation is 1.5, Construction is 1.3, unusual operation is 1.3

print(Ws, "+", Wc,"+" ,S, "/(", U, "-" , Wg, ")")
print("SFf ", round(Result, 2))
if Result >=SF:
    print("Flotation Safety Factor of ", SF, " Is met")
else:
    print("Factor of Safety for Flotation is less than ", SF ,".  SFf is not met")
```
